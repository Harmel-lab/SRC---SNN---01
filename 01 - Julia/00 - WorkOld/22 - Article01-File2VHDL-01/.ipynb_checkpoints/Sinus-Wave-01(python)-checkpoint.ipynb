```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
T = 200                  # Simulation time (ms)
dt = 1.0                 # Time step (ms)
time = np.arange(0, T, dt)
n_neurons = 5            # Number of input neurons
f_base = 0.05            # Base frequency of oscillators (Hz)
phase_shift = np.pi / 4  # Phase shift for encoding
spike_threshold = 0.9    # Arbitrary threshold for spike generation

# Generate input spike trains (random)
np.random.seed(0)
spike_trains = (np.random.rand(n_neurons, len(time)) < 0.05).astype(float)

# Generate phase oscillators
omega = 2 * np.pi * f_base  # Angular frequency
phi = np.linspace(0, 2 * np.pi, n_neurons, endpoint=False)

# Initialize holographic memory
H = np.zeros(len(time))

# Compute holographic trace
for t_idx, t in enumerate(time):
    sum_signal = 0
    for i in range(n_neurons):
        signal = spike_trains[i, t_idx] * np.sin(omega * t * dt + phi[i])
        sum_signal += signal
    H[t_idx] = sum_signal

# Generate spike output based on phase coherence (simple thresholding)
output_spikes = (H > spike_threshold).astype(float)

# Plot results
fig, axs = plt.subplots(3, 1, figsize=(10, 7), sharex=True)

# Input spikes
axs[0].imshow(spike_trains, aspect='auto', cmap='Greys', interpolation='nearest')
axs[0].set_title("Input Spike Trains")
axs[0].set_ylabel("Neuron Index")

# Holographic trace
axs[1].plot(time, H)
axs[1].axhline(spike_threshold, color='r', linestyle='--', label="Threshold")
axs[1].set_title("Holographic Trace H(t)")
axs[1].set_ylabel("Amplitude")
axs[1].legend()

# Output spikes
axs[2].stem(time, output_spikes, linefmt='g-', markerfmt='go', basefmt=" ")
axs[2].set_title("Output Spikes (Phase-Coherent Events)")
axs[2].set_xlabel("Time (ms)")
axs[2].set_ylabel("Spike")

plt.tight_layout()
plt.show()
```
