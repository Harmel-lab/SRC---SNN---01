```julia
using Random
using Plots

# ------------------
# Paramtres gnraux
# ------------------
T = 200                      # Dure de la simulation (ms)
dt = 1.0                     # Pas de temps (ms)
time = 0:dt:(T - dt)
n_neurons = 5                # Nombre de neurones d'entre
f_base = 0.05                # Frquence de base (Hz)
 = 2 * f_base              # Frquence angulaire
 = range(0, 2, length=n_neurons+1)[1:end-1]  # Phases rparties

# Seuil de dclenchement
threshold = 0.9

# ----------------------
# Gnration des spikes
# ----------------------
Random.seed!(0)
spike_trains = rand(length(time), n_neurons) .< 0.05
spike_trains = spike_trains'  # Taille : [neurones x temps]

# ----------------------
# Trace holographique H(t)
# ----------------------
H = zeros(length(time))

for (t_idx, t) in enumerate(time)
    sum_signal = 0.0
    for i in 1:n_neurons
        spike = spike_trains[i, t_idx]
        if spike == 1
            signal = sin( * t + [i])
            sum_signal += signal
        end
    end
    H[t_idx] = sum_signal
end

# ----------------------
# Spikes de sortie
# ----------------------
output_spikes = map(x -> x > threshold ? 1.0 : 0.0, H)

# ----------------------
# Visualisation
# ----------------------
default(size=(800, 600))

# 1. Spikes d'entre
p1 = heatmap(time, 1:n_neurons, spike_trains, xlabel="Temps (ms)",
    ylabel="Neurone", title="Spikes d'entre", c=cgrad([:white, :black]), colorbar=false)

# 2. Hologramme
p2 = plot(time, H, label="H(t)", xlabel="Temps (ms)", ylabel="Amplitude",
    title="Trace holographique")
hline!([threshold], l=:dash, c=:red, label="Seuil")

# 3. Spikes de sortie
#p3 = scatter(time[output_spikes .== 1.0], ones(sum(output_spikes)), 
#   xlabel="Temps (ms)", ylabel="Spike", markershape=:circle, 
#   title="Spikes de sortie (cohrence de phase)", label="", c=:green)

# Affichage combin
#plot(p1, p2, p3, layout=(3,1), linkx=true)
plot(p1, p2, layout=(2,1), linkx=true)
```
