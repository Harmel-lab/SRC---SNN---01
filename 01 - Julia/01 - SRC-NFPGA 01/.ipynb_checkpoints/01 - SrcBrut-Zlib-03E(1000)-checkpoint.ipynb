```julia
##########################################################
## Simulation d'un neurones SRC - Pascal - test FPGA    ##
##########################################################
## Modification de NeSRC3E -> FORME fpga                ##
##  Transformation de la lib NeSRC3 vers une version    ##
##  FPGA - nombre entier (16bits signé) .               ##
##                                                      ##
##                                                      ##
##                                     18 décembre 2023 ##
##########################################################

using Plots
using Printf

# utilisation de ma lib pour le modèle neuro - perso et Florent - pour mémoire 
# def using my lib -> neurone evec struct et fonction - entrées de spikes par vecteurs
# include("../MyLib/Grulib-01.jl")
include("../MyLib/FpgaLib-01.jl")

# Réalisation d'un SRC version 'Florent' avec modifications Harmeling Pacal
# suppression de l'équation zst -> fonction cycle hystéresis à seuil variable.
# modification en utilisant deux équations ReLu en place des tanh
# Harmeling - septembre 2023 -> dérnière correction le 2 octobre

#temps du calcul différentiel
const bEnd::Float32=400.0
const bStep::Float32=0.1
const TableSize::Int32 = bEnd/bStep
```

```julia
# Initialisation des vecteurs de calculs
#---------------------------------------
h=Vector{Float64}(undef,TableSize)
hs=Vector{Float64}(undef,TableSize)
zs=Vector{Float64}(undef,TableSize)
x=Vector{Float64}(undef,TableSize)
```

```julia
# utilisation de NeSRC3 avec 5 arguments - R, rs, bias, Zmaw et Zmin (Grulib-01) - Un Pulse
# -----------------------------------------------------------------------------------------
const bEnd::Float32=120.0
const bStep::Float32=0.1
const TableSize::Int32 = bEnd/bStep

h=zeros(TableSize)
hs=zeros(TableSize)
zs=zeros(TableSize)
x=zeros(TableSize)

#condition initiale
ht::Int16=0
ht1::Int16=0
hst::Int16=0
hst1::Int16=0
zst::Int16=100
xt::Int16=-1000

pointeur::Int32=1

const param=Vector{Int16}(undef,5)
param[1] = 1
param[2] = -4
param[3] = -3
param[4] = 99  # modification - permet d'adapter la fréquence des spikes -
param[5] = 0

#calcul des trois équation h,zst et hs
#le courant injecté doit être inférieure à 3
for i in 1.0:bStep:bEnd
 
    if ((i>(bEnd/3)) && (i<((bEnd/3)+50))) xt=1000 
    else xt=-1000
    end
    x[pointeur] = xt
 
    h[pointeur],hs[pointeur],zs[pointeur]=NeFsrc!(xt,ht,hst, param)

    @printf("xt=%d - h=%d - hs=%d - zs=%d\n", xt, h[pointeur], hs[pointeur] ,zs[pointeur])

    ht=h[pointeur]
    hst=hs[pointeur]
    pointeur+=1
    
end
plot(size = (900, 300))
plot!(x, xlabel="Time",label="x",  linecolor = :blue)
plot!(h,label="h",linecolor = :green)
plot!(zs,label="Z",  linecolor = :red)
plot!(hs,label="hs",linecolor = :black)

#savefig("NeSRC3E-01.png")
```
