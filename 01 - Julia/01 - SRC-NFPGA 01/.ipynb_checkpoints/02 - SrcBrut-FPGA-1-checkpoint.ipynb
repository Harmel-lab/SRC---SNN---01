```julia
##########################################################
## Simulation d'un neurones SRC - Florent et Pascal     ##
##########################################################
## Amélioration de la fonction NeSRC->NeSRC1->NeSRC2    ##
##  vers la fonction NeSRC3 qui sera plus souple dans   ##
##  la gestion de la fréquence de sortie.               ##
##  Pour y arrivé zs devient un cycle d'hystérésis      ##
##  dépendante de la valeur de ht.                      ##
##                                      24 octobre 2023 ##
##########################################################

using Plots
# utilisation de ma lib pour le modèle neuro - perso et Florent - pour mémoire 
# def using my lib -> neurone evec struct et fonction - entrées de spikes par vecteurs
include("../MyLib/FpgaLib-01.jl")

# Réalisation d'un SRC version 'Florent' avec modifications Harmeling Pacal
# suppression de l'équation zst -> fonction cycle hystéresis à seuil variable.
# modification en utilisant deux équations ReLu en place des tanh
# Harmeling - septembre 2023 -> dérnière correction le 2 octobre

#temps du calcul différentiel
const bEnd::Float32=100
const bStep::Float32=0.1
const TableSize::Float32 = bEnd/bStep
```

```julia
# utilisation de NeFsrc avec 5 arguments - R, rs, bias, Zmaw et Zmin (Fpgalib-01)
# -------------------------------------------------------------------------------

h=Vector{Float32}(undef,TableSize)
h=zeros(TableSize)

hs=Vector{Float32}(undef,TableSize)
hs=zeros(TableSize)

zs=Vector{Float32}(undef,TableSize)
zs=zeros(TableSize)

x=Vector{Float32}(undef,TableSize)
x=zeros(TableSize)

#condition initiale
ht::Float32=0.0
ht1::Float32=0.0
hst::Float32=0.0
hst1::Float32=0.0
zst::Float32=100.0
xt::Float32=-1000.0

pointeur::Int32=1

const param=Vector{Float32}(undef,5)
param[1] = 1000.0     # 
param[2] = -4000.0    # 
param[3] = -3000.0    # 
param[4] = 996.0      # modification - permet d'adapter la fréquence des spikes -
param[5] = 100.0      # 

#calcul des trois équation h,zst et hs
#le courant injecté doit être inférieure à 3
for i in 1:bStep:bEnd
 
    if ((i>10) && (i<(bEnd-10))) xt=1000.0
    else xt=-1000
    end
    x[pointeur] = xt
 
    h[pointeur],hs[pointeur],zs[pointeur]=NeFsrc!(xt,ht,hst, param)

    ht=h[pointeur]
    hst=hs[pointeur]
    pointeur+=1
    
end
plot(size = (900, 300))
plot!(x, xlabel="Time",label="x",  linecolor = :blue)
plot!(h,label="h",linecolor = :green)
plot!(zs,label="Z",  linecolor = :red)
plot!(hs,label="hs",linecolor = :black)
```
