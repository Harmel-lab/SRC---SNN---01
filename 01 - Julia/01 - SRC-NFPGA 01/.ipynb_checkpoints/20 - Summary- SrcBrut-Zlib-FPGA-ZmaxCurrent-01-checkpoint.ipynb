```julia
# -----------------------------------------------------------------------------------------------
# - Analyse du comportement SCR-VHDL suivant la valeur de Zmax et du courent d'entr√© - ANALYSE1 -
#                                                                                    - 21MARS25 -
# -----------------------------------------------------------------------------------------------
# - 1 - NeSRC3Fpga3 avec 5 arguments - R, rs, bias, Zmaw et Zmin (Fpgalib-01) (INT16)           -
# - 2 - Premiere Analyse du courant suivant Zmax=800                                            -
# - 3 - NeSRC3Fpga avec 5 arguments - R, rs, bias, Zmaw et Zmin (Fpgalib-01) (Float32)          -
# - 4 - NeSRC3Fpga3 avec 5 arguments - R, rs, bias, Zmaw et Zmin (Fpgalib-01) (INT16) S:1000    -
# -----------------------------------------------------------------------------------------------
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
# Warning - THE POSITION WHEN THE Z VALUE IS COMPUTED IS VERY IMPORTANT                         !
# Warning - The Z VALUE MANDATORY COMPUTE BEFORE HS THEREFOR HS IS 'UNSTABLE' and H OUTPUT TOO  !
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

using Plots
include("../MyLib/Srclib-01.jl")
include("../MyLib/FpgaLib-01.jl")

const bEnd::Float32=40.0
const bStep::Float32=0.1

const TableSize::Int32 = bEnd/bStep
```

```julia
# utilisation de NeSRC3Fpga3 avec 5 arguments - R, rs, bias, Zmaw et Zmin (Fpgalib-01) 
# ------------------------------------------------------------------------------------

# Initialisation des vecteurs de calculs - Forme FPGA - INT
#----------------------------------------------------------
Fh=Vector{Int16}(undef,TableSize)
Fhs=Vector{Int16}(undef,TableSize)
Fzs=Vector{Int16}(undef,TableSize)
Fx=Vector{Int16}(undef,TableSize)

Fh=zeros(TableSize)
Fhs=zeros(TableSize)
Fzs=zeros(TableSize)
Fx=zeros(TableSize)

#condition initiale
Fht::Int16=0
Fht1::Int16=0
Fhst::Int16=0
Fhst1::Int16=0
Fzst::Int16=100
Fxt::Int16=-1000

pointeur::Int32=1

const Fparam=Vector{Int16}(undef,5)
Fparam[1] = -3000
Fparam[2] = 900   # modification - permet d'adapter la fr√©quence des spikes - ATTENTION AVANT 0.99 -> Coeff. 1000 - 970 √† 999
Fparam[3] = 100

#calcul des trois √©quation h,zst et hs
#le courant inject√© doit √™tre inf√©rieure √† 3
for i in 1.0:bStep:bEnd
 
     #g√©n√©rateur courant d'entr√©e - 2 pulses - court et long
    if ((i>(bEnd/5)) && (i<((bEnd)/(5/4)))) Fxt=1000 
    else Fxt=-1000
    end
    Fx[pointeur] = Fxt
 
    Fh[pointeur],Fhs[pointeur],Fzs[pointeur]=NeSRC3Fpga4!(Fxt,Fht,Fhst, Fparam)

    Fht=Fh[pointeur]
    Fhst=Fhs[pointeur]
    pointeur+=1
    #println(Fzs)
    
end
plot(size = (1200, 400))
p1=plot(Fx, xlabel="Time",label="x",  linecolor = :blue)
p1=plot!(Fh,label="h",linecolor = :green)
p2=plot(Fx, xlabel="Time",label="x",  linecolor = :blue)
p2=plot!(Fzs,label="Z",  linecolor = :red)
p3=plot(Fx, xlabel="Time",label="x",  linecolor = :blue)
p3=plot!(Fhs,label="hs",linecolor = :black)
plot!(p2,p1,p3, layout = (3, 1))

#savefig("NeSRC3Fpga4 (l3-1_Z950).pdf")
```
